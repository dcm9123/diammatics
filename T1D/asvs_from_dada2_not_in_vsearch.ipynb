```python
import os
import pandas as pd
```

```python
#I manually added column names to the vsearch file
#Fetching the unidentified ASVs by VSearch
df_vsearch = pd.read_csv("output/vsearch_taxonomy_w9w10_output_raw_993.tsv", sep='\t')
df_vsearch
unidentified_asvs = df_vsearch.loc[df_vsearch['taxonomy']=='*',['asv_id','taxonomy']] #Adding column names manually
unidentified_asvs_list = unidentified_asvs['asv_id'].tolist() #The star from the line above indicates no taxonomy identified
```

```python
#Fetching the corresponding DADA2 IDs
matching_rows =[]
df_dada2 = pd.read_csv("output/dada2_assignTaxonomy_w9w10_rdp.csv")
dada2_asv_id = df_dada2['ASV_ID'].tolist()
for item in unidentified_asvs_list:
    if item in dada2_asv_id:
        index = dada2_asv_id.index(item)
        matching_rows.append(df_dada2.iloc[index])

other_asvs = pd.DataFrame(matching_rows)
```

```python
other_asvs.to_csv('filtered_dada2_asv_w9w10.csv')
```

```python
unidientified_asvs_by_dada2 = other_asvs[other_asvs['Genus'].isna()]
unidientified_asvs_by_dada2_and_id = unidientified_asvs_by_dada2['ASV_ID']
unidientified_asvs_by_dada2_and_id.to_csv('unidentified_asvs_by_dada2_and_vsearch_w9w10.csv')
```
