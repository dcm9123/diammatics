```python
import pandas as pd
import os
```

```python
df_metadata = pd.read_csv("w9w10_subset_metadata.csv",sep=",")
df_picrust2 = pd.read_csv("EC_metagenome_out/pred_metagenome_unstrat.tsv",sep="\t", header=None)
df_picrust2 = df_picrust2.T
df_picrust2.columns = df_picrust2.iloc[0] #giving the first row the header 
df_picrust2 = df_picrust2[1:]
df_picrust2['function'] = df_picrust2['function'].str.lstrip('X')
merged_df = pd.merge(df_picrust2, df_metadata[['Sequence ID', 'Diabetic', 'FatePrimary', 'Sex', 'Human_derived_status']], left_on='function', right_on = 'Sequence ID', how='left')
merged_df.to_csv('merged_EC_file_by_metadata.csv', index=False)
```
