```python
#importing packages
import pandas as pd
import os
from Bio import SeqIO
```

```python
#Defining files
master_file = pd.read_csv("WGS_all_runs/FINAL_report_WGS021224.csv")
fasta_file = "16S_data/16S_from_both/all_approaches_merged_rmdup.fasta"
```

```python
#This sections corresponds to the first DADA2 file that needs the whole taxonomy as part of its header
master_file_filtered = master_file.loc[master_file['Selected for Downstream']=='Yes',
                                       ['Sample number','Classification']] #filters selected for downstream and prints these two columns only into a new df
master_file_filtered['Sample number'] = master_file_filtered['Sample number'].apply(lambda x: f'{int(x):03d}') #This ensures that I have always three digits
sample_classification_list = master_file_filtered['Classification'].tolist() #transforms column to list
sample_number_list = master_file_filtered['Sample number'].tolist() #transforms colum to list
updated_records = [] #creates an empty list of updated records to be added (in this case updated header)
with open(fasta_file,'r') as fasta:
    for record in SeqIO.parse(fasta,"fasta"):
        header = record.id
        sequence = str(record.seq)
        header_sample_number = header.split("_")[0] #makes sure I only get the first part '>seqxxx'
        header_sample_number = header_sample_number.split("seq")[1] #makes sure I only get the number 'xxx'
        if header_sample_number in sample_number_list: #if the header from the original fasta file is in the excel file then do the following
            index = sample_number_list.index(header_sample_number) #saves the index number
            new_header = f"{sample_classification_list[index]}" #gets the classification corresponding to the index number
            record.id = new_header #updates new header to the existing record
        updated_records.append(record)

with open("dada2_new_updatad_taxonomy.fasta","w") as updated_fasta:
    SeqIO.write(updated_records,updated_fasta,"fasta") #writes new updated fasta file
```

```python
#This section corresponds to the second DADA2 file that needs just the ID Genus Species for the 'AssignSpecies'
#code
master_file_filtered2 = master_file.loc[master_file['Selected for Downstream']=='Yes',
                                        ['Sample number','GTDBtk Species Classification (pplacer)']]
master_file_filtered2['Sample number'] = master_file_filtered2['Sample number'].apply(lambda x: f'{int(x):03d}') #This ensures that I have always three digits
sample_classification_list2 = master_file_filtered2['GTDBtk Species Classification (pplacer)'].tolist()
sample_number_list2 = master_file_filtered2['Sample number'].tolist()
updated_records2 = []
with open(fasta_file,'r') as fasta:
    for record2 in SeqIO.parse(fasta, "fasta"):
        header2 = record2.id
        sequence2 = str(record2.seq)
        header_sample_number2 = header2.split("_")[0]
        header_sample_number2 = header_sample_number2.split("seq")[1]
        if header_sample_number2 in sample_number_list2:
            index2 = sample_number_list2.index(header_sample_number2)
            new_header2 = f"{header2} {sample_classification_list2[index2]}"
            record2.id = new_header2
            print(header2 +" "+ sample_classification_list2[index2])
        updated_records2.append(record2)
print(updated_records2[0])
with open("dada2_assign_species.fasta","w") as updated_fasta2:
    SeqIO.write(updated_records2, updated_fasta2, "fasta")
```
